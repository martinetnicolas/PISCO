```python
%pylab inline
import pisco
import pisco.datasets
import tensorflow_datasets as tfds
```

```python
dset = tfds.load('pisco_euclid',split='train')
dset = dset.as_numpy_iterator()
```

```python
# Grab one example from the dataset
for i in range(2):
    figure()
    example = next(dset)
    imshow(example['image'],cmap='gray_r')
```

```python
dset = tfds.load('pisco_lsst',split='train')
dset = dset.as_numpy_iterator()
```

```python
from astropy.visualization import make_lupton_rgb
def luptonize(img):
  return make_lupton_rgb(img[:,:,2], img[:,:,1], img[:,:,0],
                         Q=15, stretch=0.5, minimum=0)
```

```python
# Grab one example from the dataset
example = next(dset)
imshow(luptonize(example['image']))
title('Shear: g1 = %0.2f g2 = %0.2f'%(example['g'][0], example['g'][1]));
```
